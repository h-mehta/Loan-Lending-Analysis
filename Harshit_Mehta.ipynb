## Objective:
To identify the risky loan applicants at the time of loan application so that such loans can be reduced thereby cutting down the amount of credit loss. Identification of such applicants using EDA is the aim of this case study.

Our Hypothesis would be : <br><br>
<b> Null Hypothesis :</b> Client will default on the loan
<br><b> Alternate Hyp : </b>Client will not default

We are trying to reduce the chances of :

1. giving loans to customers who will default ( Type I error )
2. not giving loans to cutomers who will repay the loan ( Type II error )

Let's consider other variables,

Does interest rate influence the customer's likelihood to default? Let's find out,

```python
master.int_rate.describe()
```

Let's divide the rows into slabs based on the 4 Quartiles -

```python
def get_interest_slab(x):
    if x>5.0 and x<=9.32:
        return "Slab 1"
    elif x>9.32 and x<=11.86:
        return "Slab 2"
    elif x>11.86 and x<=14.65:
        return "Slab 3"
    else:
        return "Slab 4"
```

```python
# Adding a derived matric to the the data
master["interest_slab"] = master.int_rate.apply(get_interest_slab)
master.head()
```

```python
D_interest_slab = master.pivot_table(values = 'loan_amnt', index = ['interest_slab','loan_status'], aggfunc = ['sum','count','median'])
D_interest_slab.columns = ["Sum","Count","Median"]
D_interest_slab
```

```python
D_interest_slab = D_interest_slab.reset_index()
D_interest_slab
```

```python
DI_1 = D_interest_slab.loc[D_interest_slab["interest_slab"]=="Slab 1",:]
DI_2 = D_interest_slab.loc[D_interest_slab["interest_slab"]=="Slab 2",:]
DI_3 = D_interest_slab.loc[D_interest_slab["interest_slab"]=="Slab 3",:]
DI_4 = D_interest_slab.loc[D_interest_slab["interest_slab"]=="Slab 4",:]
```

```python
# Loan status wise split based on interest_slab
#figure having 4 subplots

fig, axs = plt.subplots(2,2, figsize=(16,15))

# subplot 1
axs[0,0].pie(list(DI_1["Count"]), labels = list(DI_1["loan_status"]),autopct='%1.2f%%') 
axs[0,0].set_title(" Loan status for Slab 1 interest loans ", bbox={'facecolor':'0.8', 'pad':5})

# subplot 2 
axs[0,1].pie(list(DI_2["Count"]), labels = list(DI_2["loan_status"]),autopct='%1.2f%%')
axs[0,1].set_title(" Loan status for Slab 2 interest loans ", bbox={'facecolor':'0.8', 'pad':5})

# subplot 3
axs[1,0].pie(list(DI_3["Count"]), labels = list(DI_3["loan_status"]),autopct='%1.2f%%') 
axs[1,0].set_title(" Loan status for Slab 3 interest loans ", bbox={'facecolor':'0.8', 'pad':5})

# subplot 4 
axs[1,1].pie(list(DI_4["Count"]), labels = list(DI_4["loan_status"]),autopct='%1.2f%%')
axs[1,1].set_title(" Loan status for Slab 4 interest loans ", bbox={'facecolor':'0.8', 'pad':5})

```

```python
D_issue_d_month = master.pivot_table(values = 'loan_amnt', index = ['issue_d_month'], aggfunc = ['count','median'])
D_issue_d_month.columns = ["Count","Median"]
D_issue_d_month = D_issue_d_month.reset_index()
D_issue_d_month 
```

Let's divide the year into 4 quater's - would be deriving the quater from month column,

```python
def get_quater(x):
    if x>=1 and x<=3:
        return "Q1"
    elif x>=4 and x<=6:
        return "Q2"
    elif x>=7 and x<=9:
        return "Q3"
    else:
        return "Q4"
```

```python
D_issue_d_month["quater"] = D_issue_d_month["issue_d_month"].apply(get_quater)
D_issue_d_month
```

```python
D_issue_quater = D_issue_d_month.pivot_table(values = 'Median', index = ['quater'], aggfunc = ['mean'])
D_issue_quater.columns = ["Mean"]
D_issue_quater["Mean"] = D_issue_quater["Mean"].apply(lambda x:round(x,1))
D_issue_quater = D_issue_quater.reset_index()
D_issue_quater
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
y_axis_data = list(D_issue_quater["Mean"].apply(lambda x:x/10000))
ax.bar( list(D_issue_quater["quater"]) , y_axis_data )
ax.set_xlabel('Quater')
ax.set_ylabel('Average value of Loan Amount (in ten thousands)')
ax.set_title('Seasonal variation on Loan Amount')
plt.show()
```

**Inference 8** :- We can see that the bank is more likely to sanction a higher value loan around New Year, i.e. Q1 and Q4

But why does the bank lend more during some quaters? Is the bank more likely to sanction loan in a particular quater OR can we attribute the above observation to customer behaviour? Let's check,

```python
D_month_purpose = master.pivot_table(values = 'loan_amnt', index = ['issue_d_month','purpose'], aggfunc = ['sum','count','median'])
D_month_purpose.columns = ["Sum","Count","Median"]
D_month_purpose.sort_values(by = ["issue_d_month","Count"],ascending = False)
```

What was the most dominant reason for loans?

```python
dmp = D_month_purpose.sort_values(by = ["issue_d_month","Count"],ascending = False)
dmp = dmp.reset_index()
dmp
```

Is this trend followed in every quater?

```python
dmp["quater"] = dmp["issue_d_month"].apply(get_quater)
dmp
```

```python
dmp = dmp.pivot_table(values = 'Count', index = ['quater','purpose'], aggfunc = ['sum'])
dmp.columns = ["Count"]
dmp.sort_values(by = ["quater","Count"],ascending = False)
```

**Inference 9** :- Since most dominant purpose of loans across all quaters is "Debt consolidation" and "Credit_Card", it can be inferred that due to higher spending around the Christmas/New year break, there is a spike credit card usage / requirement for loans.
Thus, it is not the case that the bank tends to clear more loans in Q1 and Q4, but because there is a higher demand from clients during the Q1 and Q4 season, we observe more loans being sanctioned.
